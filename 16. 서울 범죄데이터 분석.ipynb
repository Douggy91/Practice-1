# 연습문제 
1. 살인발생이 가장 높은 관서명, 살인발생, 살인검거를 출력하시오
2. 관서 별 절도발생, 절도검거에 대한 바 차트를 그리시오
3. 절도검거율 컬럼을 추가하시오

```python
df.loc[df['살인 발생'].idxmax(), ['관서명','살인 발생','살인 검거']]
```

```python
df.set_index('관서명', inplace = True)
mx = df['살인 발생'].max()
df[df['살인 발생'] == mx][['살인 발생','살인 검거']]
```

```python
df1 = df[['절도 발생', '절도 검거']]
```

```python
df1.plot(kind ='bar', figsize = (10,8))

plt.show()
```

```python
df['절도 검거율'] = round(df['절도 검거']/df['절도 발생'],2)
df
```
